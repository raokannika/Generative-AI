### Wordnet lemmatizer
Lemmatization technique is like stemming. The output we get after lemmatization is called 'lemma' =, which is the root word rather than root stem, which is the output of stemming. After lemmatization we will be gettign valid word which means the same thing, doesn't change the meaning of the word

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer= WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going")
```

```python
# two functions of lemmatization are (word, post tag)
# post tags
# Noun-n
# verb= v
# adjective= a
# adverb= r

#verb is highly preffered and important for lemmatization
lemmatizer.lemmatize("going", pos="v")
```

```python
lemmatizer.lemmatize("going", pos="n")
```

```python
lemmatizer.lemmatize("going", pos="a")
```

```python
lemmatizer.lemmatize("going", pos="r")
```

```python
words = ['eaten', 'eating', 'eats', 'writing', 'writes', 'programming', 'programs', 'finalize', 'finalizing', 'history']
```

```python
for word in words:
    print(f"{word} ----> {lemmatizer.lemmatize(word, pos='v')}")
```
