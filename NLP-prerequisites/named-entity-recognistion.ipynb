```python
sentence = "The Eiffel Towe was built from 1887 to 1889 by French Engineer Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
words = nltk.word_tokenize(sentence)
```

```python
# NLTK data is already downloaded above.
```

```python
tag_elements = nltk.pos_tag(words)
```

```python
nltk.ne_chunk(tag_elements)
```
