```python
corpus = """Hello welcome, to the world of Natural Language Processing.
Natural Language Processing focuses on the interaction between computers and humans through natural language. 
It enables machines to understand, interpret, and generate human language in a valuable way. 
hello! kannika
"""
```

```python
print(corpus)
```

```python
## Tokenization
from nltk.tokenize import sent_tokenize 
## sent_tokenize is a function used to split a paragraph (or large text) into individual sentences.
```

```python
sent_tokenize(corpus )
```

```python
from nltk.tokenize import word_tokenize 
```

```python
word_tokenize(corpus)
```

```python
# one more library which makes sure that punctuation is also treated as a separate token
# like ' "" etc
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
# tree bank tokenizer- full stop will not be treated as separate token
# byt with respect to last full stop in the corpus, it is treated as a separate token
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
#you can also use this method
#TreebankWordTokenizer().tokenize(corpus)
```

```python
tokenizer.tokenize(corpus)
```
