### Stemming
Is the process of reducing word to its word stem that affixes to suffixes and prefixes or to the root of words known as lemma.

## For example
if a review has words like eat, eating,eaten the stem for for those is just eat
going,goes,gone--> go

```python
words = ['eaten', 'eating', 'eats', 'writing', 'writes', 'programming', 'programs', 'finalize', 'finalizing', 'history']
```

```python
# PorterStemmer

from nltk.stem import PorterStemmer
```

```python
stemming= PorterStemmer()
```

```python
for word in words:
    print(word+"---->"+stemming.stem(word))
```

for history the stem word turns out to be histori
this raises one disadvantage of stemming, for some of the words correct meaning is not found

```python
stemming.stem('Congratulations')
```

```python
stemming.stem('sitting')
```

### RegexpStemmer Class
NLTK has RegexpStemmer Class with the help of which we can easily implement Regular Expression stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer()
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
# dollar at the end of the pattern means that the pattern should be at the end of the word
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

### Snowball stemming


```python
from nltk.stem import SnowballStemmer 
```

```python
snowballstemmer=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"---->"+snowballstemmer.stem(word))
```
