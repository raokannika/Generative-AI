### Stemming
Is the process of reducing word to its word stem that affixes to suffixes and prefixes or to the root of words known as lemma.

## For example
if a review has words like eat, eating,eaten the stem for for those is just eat
going,goes,gone--> go

```python
words = ['eaten', 'eating', 'eats', 'writing', 'writes', 'programming', 'programs', 'finalize', 'finalizing', 'history']
```

```python
# PorterStemmer

from nltk.stem import PorterStemmer
```

```python
stemming= PorterStemmer()
```

```python
for word in words:
    print(word+"---->"+stemming.stem(word))
```

for history the stem word turns out to be histori
this raises one disadvantage of stemming, for some of the words correct meaning is not found
