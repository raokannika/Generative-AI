Dictonaries are unorderes collections of items. They store data in key-value pairs. 
Keys must be unique and immutable(example: strings, numbers, or tuples), while values can be of any type

```python
##Creating Dictionaries
## [] for list , () for tuples and for dictinaries it is {}
empty_dict = {} # creates empty dicinaries
print(type(empty_dict))
```

```python
## Another way to create empty dictinaries is initializin with method
empty_dict=dict()
print(empty_dict)
```

```python
student = {"name":"Kannika", "age" : 20 , "grade": 'A'}
print(student)
print(type(student))
```

```python
## Possibility of getting not expected output
## when the key is not unique
student = {"name":"Kannika", "age" : 20 , "name": 'A'}
print(student)
```

accessing dictionary elements

```python
student = {"name":"Kannika", "age" : 20 , "grade": 'A'}
print(student)
```

```python
print(student('grade'))
```

```python
print(student['grade'])
print(student['age'])
```

Another way of acessing dictionary elements using get() method

```python
print(student.get('grade'))
print(student.get('age'))
# print(student.get('last_name'))
#to see how it will give output when the key not is not present and is trying to access

#giving default value
print(student.get('last_name',"Not Available"))
```

Modifying dictionary elements
Dictionary is mutable, so you can add, update or delete elements

```python
print(student)
```

```python
student["age"]=21 # updated value for age
print(student)

student["addresss"] = "India" # added a new key and value
print(student)
```

```python
#to delete a key and value pair
del student['grade']

print(student)
```
