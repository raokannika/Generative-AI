Dictonaries are unorderes collections of items. They store data in key-value pairs. 
Keys must be unique and immutable(example: strings, numbers, or tuples), while values can be of any type

```python
##Creating Dictionaries
## [] for list , () for tuples and for dictinaries it is {}
empty_dict = {} # creates empty dicinaries
print(type(empty_dict))
```

```python
## Another way to create empty dictinaries is initializin with method
empty_dict=dict()
print(empty_dict)
```

```python
student = {"name":"Kannika", "age" : 20 , "grade": 'A'}
print(student)
print(type(student))
```

Modifying dictionary elements
Dictionary is mutable, so you can add, update or delete elements

```python
print(student)
```

```python
#to print keys and values
key=student.keys()
print(key)
values=student.values()
print(values)
# to get key value pairs as list of tuples like it will look like
#[(key,value),(key,value)]
items=student.items()
print(items)
```

```python
# to understand the need of shallow copy
student_copy = student
print(student)
print(student_copy)
```

```python
student["name"]="kannika change"
print(student)
print(student_copy)
# here we can see that when changes are made in the main copy it is affected in the copied dictonaries as well
# that is not how copy of data to other should be working
# that's why the concept of shallow copy comes
# if we want to save a dictonaries as copy and make changes in the copy/orginal
#  that will not affect the other we will have to use shallow copy
```

```python
# shallow copy inturns provides different memory locations
student_copy1= student.copy()
print(student_copy1)
print(student)
```

```python
student["name"] = "Kannika"
print(student)
print(student_copy1)
```

```python
#iterating over dicionaries
# loops can be used to iterate over dictonaries, key, values or items

# Iterating over key
for keys in student.keys():
    print(keys)
```

```python
# Iterating over value
for value in student.values():
    print(value)
```

```python
# Iterating over key value pairs
for key,value in student.items():
    print(f"{key} : {value}")
```

```python
# Nested dictinaries 
students = {
    "student1":{"name":"kannika","age":20},
    "student2":{"name":"Ashitha","age":25},
    "student3":{"name":"Akshara","age":1}
}
print(students)
```

```python
# Access nested dictonaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
# Iterating over nested dictonaries

for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
# Dictionary comphrehension
squares = {x:x**2 for x in range(5)}
print(squares)
```

```python
# condition dictinary comprehension
evens = {x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
# use a dictionary to count the frequency of elements in list
numbers={1,2,2,3,3,4,4,4,4}
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number]=1

print(frequency)
```

```python
## Merge two diconaries into one
dict1={"a":1,"b":2}
dict2={"b":3,"c":4}

merged_dict={**dict1,**dict2}
print(merged_dict)
```
