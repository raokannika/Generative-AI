### Immutable Nature of tuple
Meaning : tuple elements cannot be changed once assigned

```python
# giving an example to show mutability
lst=[1,2,3,4,5]
print(lst)

lst[1]="Bob"
print(lst)
```

### Packing and Unpacking Tuple

```python
# Whatever things are assigned here will be packed into a tuple
packed_tuple=1,"hello",3.14
print(packed_tuple)
```

```python
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
## unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested tuple

## Nested list
lst=[[1,2,3,4],[6,7,8,9],(1,"hello","world",3.14,"c")] ## we can have tuples inside list
print(lst)
lst[0]
lst[0][0:3]
lst[2][0:3]
```

```python
nested_tuple=((1,2,3),("a","b","c"),(True, False))
print(nested_tuple)

## Accessing the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
