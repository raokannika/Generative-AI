```python
def temp_con(temp, tempType):
    if tempType == "C":
        return (temp * 9/5) + 32
    elif tempType == "F":
        return (temp - 32) * 5/9
    else:
        return None

tempType = input("Enter the unit (C or F): ")
temp = float(input("Enter the temperature: "))

converted = temp_con(temp, tempType)

if converted is not None:
    print("Converted temperature: %.2f" % converted)
else:
    print("Invalid unit entered.")
```

```python
def is_strong_password(password):
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in "!@#$%^&*()_+" for char in password):
        return False
    
    return True


#Call
password = input("Enter the password: ")
fun=is_strong_password(password)
print("Is the password strong: ",fun)
```

```python
def calculate_total_cost(cart):
    totalCost=0
    for item in cart:
        totalCost += item['price']*item['quantity']
    return totalCost

cart=[
    {'name':'Apple','price':100,'quantity':6},
    {'name':'Banana','price':70,'quantity':5},
    {'name':'Dragon Fruit','price':150,'quantity':2}
]
price=calculate_total_cost(cart)
print("Total Cost : %.2f" % price)
```

```python
def is_palindrome(s):
 s=s.lower().replace(" ","")
    return s==s[::-1]

string=input("Enter a string : ")
palindrome=is_palindrome(string)
print("Is the string Palindrom: ",palindrome)
```

```python
def factorial(n):
    if n==0 :
        return 1
    else:
        return n * factorial(n-1)
    
num=int(input("Enter a number: "))
fact=factorial(num)
print("Factorial is: ",fact)
```

```python
import re
print("Re working")
def is_valid_email(email):
    pattern= r'^[a-zA-Z0-9_,+-]+@[a-zA-Z0-9-.]+$'
    return re.match(pattern,email) is not None


email=input("Enter the email: ")
test=is_valid_email(email)
print("Is the email valid? ",test)
```
