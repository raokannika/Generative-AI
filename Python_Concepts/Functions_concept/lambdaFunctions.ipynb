### Lambda function
Small anonymous functions defined using the lambda function. They can have number of arguments but only one expression. Commonly used for short operations or as arguments to higher-order function.

## Syntax
lambda arguments : expression

```python
# Normal fucntion
def add(a, b):
    return a + b

add(2, 3)  # Output: 5
```

```python
# Lambda function
addition = lambda a, b: a + b
print(addition(2, 3))
```

few more examples

```python
def even(num):
    if (num % 2 == 0):
        return True
    else:
        return False

print("4 is even? ",even(4))  # Output: True
print("5 is even? ",even(5))  # Output: False
```

```python
even1 = lambda num: num % 2 == 0

print("12 is even? ",even1(12))
```
