```python
#why fucntions?
# if we are asked to check if a number is even or odd we need write the if else statement again and again.
# Instead we can just write a function and use it for multiple numbers
def even_odd(num):
    """This function is to check if a number is odd or even"""
    if num%2==0 :
        print("The number is even")
    else :
        print("The number is odd")
```

```python
# to call the function
even_odd(24)
```

```python
# function with multiple parameters
def add(a,b):
    return a+b

result = add(2,4)
print(f"Sum:{result}")
```

```python
# Defaukt parameters
def greet(name):
    print(f"Hello {name} welcome to the page")
greet()
```

```python
def greet(name="Guest"):
    print(f"Hello {name} welcome to the page")
greet()
greet("Kannika")
```

```python
# Variable length arguments
# positional and keywords arguments
def print_numbers(*args): #positional arguments
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"john")
```

```python
# Keyword arguments
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Kannika", age="32", country="India")
```

```python
def print_details1(*args,**kwargs):
    for val in args:
        print(f"Positional argument : {val}")
    print("\nKeyword Arguments")
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details1(1,2,3,4,"kannika", name="akshara",age="12", country ="India")
```
