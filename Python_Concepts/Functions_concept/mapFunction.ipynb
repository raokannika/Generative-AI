### Defination
the map() function applies a given function to all items in an input list(or any other iterable) and returns a map object(iterator). 
Particularly used for transforming data in a list comprehensively
## Syntax:
map(funcition, *iterables)

```python
def square(x):
    return x * x
square(5)
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
map(square, numbers)
## output shows that map object has been created at the location mentioned in the memory
```

```python
# to get all the values in the form of list we execute the code- we call list() function
list(map(square, numbers))
```

```python
## Lambda function with map
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
list(map(lambda x: x * x, numbers))
```

```python
# can we map multiple iterable?
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
added_numbers = list(map(lambda x,y: x+y,numbers1, numbers2))
print(added_numbers)
```

```python
## map() to convert a list of strings to integers
str_numbers = ['1', '2', '3', '4', '5']
int_numbers = list(map(int, str_numbers))
print(int_numbers) # output: [1, 2, 3, 4, 5]
```

```python
# converting list of strings to uppercase using .upper() method with map()
words = ['apple' , 'banana' , 'cherry']
upper_word= list(map(str.upper, words))
print(upper_word)
```

```python
def get_name(person):
    return person['name']

people = [
    {'name':'Kannika' , 'age' : 20},
    {'name':'Ashitha' , 'age' : 26},
    {'name':'Ashalatha' , 'age' : 55}
]
list(map(get_name, people))
```
