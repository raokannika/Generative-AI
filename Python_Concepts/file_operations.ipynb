### File operation- Read and Write Files
File handling is a crucial part of programming lang. Python provides built in functions and methods to read from and write to files, both tet and binary.

```python
#Read a whole file
# open('file name', 'operation/mode')
#.read() method reads the entire content of the file
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# Read a file line by line r- read mode
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# To read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line)
```

```python
#  .strip() removes new line characters
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
# Writing a file (Overwriting) w- write mode

with open('example.txt', 'w') as file:
    file.write('Hello World!\n')
    file.write('This is a new line!\n')
```

```python
# Write a file without overwriting a- append mode
with open('example.txt', 'a') as file:
    file.write('Append operation is taking place.\n')
```

```python
with open('example.txt','r') as file:
    for line in file:
        print(line.strip())
```

```python
# wiritng a list of lines to a file
# writelines() method writes a list of lines to a file 
lines=['First Line\n', 'Second Line\n', 'Third Line\n']
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

### Binary files

```python
# writing into a binary file
# wb- write byte
data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
# Read the content from a source text file and write to a destination text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as dest_file:
    dest_file.write(content)
```

```python
# Counting lines,words and characters in a file
def count_test_file(file_path):
    with open('example.txt', 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count
file_path = 'example.txt'
lines, words, characters = count_test_file(file_path)
print(f'Lines: {lines},Words: {words},Characters: {characters}')
```

w+ mode - in python it is used to open a file for both reading and writing. If the file does not exist, it will be crrated. If the file exists, its content is truncated(i.e the file is overwritten)

```python
# Writing and then reading a file

with open('example.txt', 'w') as file:
    file.write('Hello World!\n')
    file.write('This is a new line!\n')

    
    file.seek(0)
    # read the content of the file
    content = file.read()
    print(content)
```

```python
with open('example.txt', 'w+') as file:
    file.write('Hello World!\n')
    file.write('This is a new line!\n')
    
    # make sure to move the file coursor to the beginning 
    # using file.seek
    file.seek(0)  # move cursor to beginning
    # read the content of the file
    content = file.read()
    print(content)
```
